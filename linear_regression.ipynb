```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import random
```

```python
def regression_function(samples=100) -> ([int], [float]):
    X = []
    Y = []
    for x in range(samples):
        # y = 5 * x + 1
        y = 5 * x + random.uniform(-100, 100)
        X.append(x)
        Y.append(y)

    return X, Y
```

```python
def print_function(X: [int], Y: [float]):
    fig, ax = plt.subplots()
    ax.scatter(X, Y)
    plt.show()
```

```python
X, Y = regression_function()
print_function(X, Y)
```

```python
def print_prediction_function(X: [int], Y: [float], Y_predicted: [float]):
    fig, ax = plt.subplots()
    ax.scatter(X, Y)
    ax.plot(X, Y_predicted, 'r')
    plt.show()
```

```python
from perceptron.linear_regression import NeuralNetwork

nn = NeuralNetwork()
Y_predicted = nn.predict(X)
print_prediction_function(X, Y, Y_predicted)
```
